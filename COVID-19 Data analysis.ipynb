### DATA is "same", "inference" makes the difference. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corona_data_csv=pd.read_csv(r'E:\Mohit\Mohit Projects\COVID 19 data analysis using python\Dataset\covid19_Confirmed_dataset.csv')
corona_data_csv.head()
```

```python
corona_data_csv.shape
```

```python
# lets delete the useless columns
df = corona_data_csv.drop(['Lat','Long'],axis=1)
# axis=0 is used for the row deletion.
df.head(10)
```

```python
# lets delete the useless columns, but this time this will effet the main dataframe...
corona_data_csv.drop(['Lat','Long'],axis=1,inplace=True)
# axis=0 is used for the row deletion.
corona_data_csv.head(10)
```

```python
# aggregating all the rows with country region...\
corona_data_aggregated=corona_data_csv.groupby('Country/Region').sum()
corona_data_aggregated.head(10)
```

```python
corona_data_aggregated.shape
```

```python
# data visualization...
corona_data_aggregated.loc['China'].plot()
corona_data_aggregated.loc['Italy'].plot()
#corona_data_aggregated.loc['India'].plot()
plt.legend()
```

```python
corona_data_aggregated.loc['China'][:3].plot()
```

```python
# first derivate of the curve...
corona_data_aggregated.loc['China'].diff().plot()
```

```python
# maximum infected rate change day by day...
corona_data_aggregated.loc['China'].diff().max()
# this shows maximum increment in the no. of corona cases.
```

```python
corona_data_aggregated.loc['Italy'].diff().max()
```

```python
# lets find the maximum infection rate for all the countries...
countries = list(corona_data_aggregated.index)
maximum_infection_rates=[]
for country in countries:
    maximum_infection_rates.append(corona_data_aggregated.loc[country].diff().max())
maximum_infection_rates
```

```python
corona_data_aggregated["maximum_infection_rate"]=maximum_infection_rates
```

```python
corona_data_aggregated.head()
```

```python
# most important dataframe...
corona_data=pd.DataFrame(corona_data_aggregated['maximum_infection_rate'])
corona_data.head()
```

```python
# Lets analyse the world happiness report now...
happiness_report=pd.read_csv(r'E:\Mohit\Mohit Projects\COVID 19 data analysis using python\Dataset\worldwide_happiness_report.csv')
happiness_report.head()
```

```python
happiness_report.shape
```

```python
# drop the useless coulumns now...
useless_col=['Overall rank','Score','Generosity','Perceptions of corruption']
happiness_report.drop(useless_col,axis=1,inplace= True)
happiness_report.head()
```

```python
happiness_report.shape
```

```python
#df= happiness_report.groupby(['Country or region'])
#df.head()
happiness_report.set_index("Country or region",inplace= True)
happiness_report.head()
```
